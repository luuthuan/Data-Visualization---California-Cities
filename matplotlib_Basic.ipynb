## `Pyplot API` and `Object-Oriented (OO) API`
* `Pyplot API` -> Quickly
* `OO API` -> Advanced

### 1. Pyplot API


```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x), color='green', linestyle='dashed', label='sin(x)');
plt.plot(x, np.cos(x), color='blue', linestyle='solid', label='cos(x)');

plt.title('Sin(x) and Cos(x)'); # title
plt.xlabel('X'); # X label
plt.ylabel('F(x)'); # Y label
# plt.xlim(0,4); # X limit
# plt.ylim(-0.5,0.5); # Y limit
# plt.axis([0,4,-0.5,0.5]); # Xmin, Xmax, Ymin, Ymax
plt.axis('equal')


plt.legend(); # Use this function to show label
```

### 2. Object-Oriented API

```python
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5)) # Figure size = Width & Height of the plot
# fig: là bức hình tổng
# ax: là các bức nhỏ
# sử dụng ax để vẽ đồ thị

#3. Plot the data
ax.plot(x,y, color='green');

ax.set(title='A simple plot', xlabel='X-axis', ylabel='Y-axis');
```

## 3. Most common types of Matplotlib plots
* `line`
* `scatter`
* `bar`
* `histogram`
* `subplots`

#### 3.1 Line

```python
# Create an array
x = np.linspace(0,10,100)
x[:5]
```

```python
# OO API plot a line chart
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,x**3);
ax.set(title='Line Plot')
```

#### 3.2 Scatter

##### 3.2.1 Scatter with Pyplot API

```python
# Pyplot API
plt.scatter(x, np.exp(x));
```

##### 3.2.2 Scatter with OO API

```python
# OO API

# Prepare New Data
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

sizes = 1000*rng.rand(100)
colors = rng.rand(100)

fig, ax = plt.subplots(figsize=(5,5))
img = ax.scatter(x,y, s=sizes, c=colors, alpha=0.4);

fig.colorbar(img);
```

#### 3.3 Bar
* Vertical (cột)
* Horizontal (dòng)

##### 3.3.1 Vertical Bar

```python
# Prepare Data from a python dictionary
soft_drinks_prices = {'Cocacola': 10, 'Pepsi': 15, 'Fanta': 20}

fig,ax = plt.subplots(figsize=(5,5))
ax.bar(soft_drinks_prices.keys(), soft_drinks_prices.values());
ax.set(title='Soft Drink Price', ylabel='Price ($)')
```

##### 3.3.2 Horizontal Bar

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.barh(list(soft_drinks_prices.keys()), list(soft_drinks_prices.values()));
ax.set(title='Soft Drink Price', xlabel='Price ($)')
```

#### 3.4 Histogram - Biểu đồ thể hiện tần suất dưới dạng cột

```python
# Prepare Data
np.random.seed(42)
student_height = np.random.normal(170, 10, 250) # chiều cao trung bình, độ lệch chuẩn, số lượng học sinh
```

```python
student_height[:5]
```

##### 3.4.1 Pyplot API

```python
# Pyplot API
plt.hist(student_height, bins=50);
```

##### 3.4.2 OO API

```python
# OO API
fig, ax = plt.subplots(figsize=(5,5))
ax.hist(student_height, bins=50);
```

#### 3.5 Subplots

##### 3.5.1 Option 1

```python
# Option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot data to each axis
ax1.plot(x,x/2);
ax1.plot(x,x*2);

ax2.scatter(np.random.random(20), np.random.random(20));
ax3.bar(soft_drinks_prices.keys(), soft_drinks_prices.values());
ax4.hist(student_height, bins=50);
```

##### 3.5.2 Option 2

```python
# Option 2:
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot data to each axis
ax[0,0].plot(x,x/2); # row 1 and col 1
ax[0,0].plot(x,x*2); # row 1 and col 1

ax[0,1].scatter(np.random.random(20), np.random.random(20)); # row 1 and col 2
ax[1,0].bar(soft_drinks_prices.keys(), soft_drinks_prices.values()); # row 2 and col 1
ax[1,1].hist(student_height, bins=50); # row 2 and col 2
```

## Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('/content/drive/MyDrive/Colab datasets/California Cities/california_cities.csv')
cities.head()
```

```python
# Extract latd and longd
lat, long = cities['latd'], cities['longd']

# Extract population and area total km2
population, area = cities['population_total'], cities['area_total_km2']
```

```python
population[:10]
```

```python
# Plot using Pyplot API
plt.figure(figsize=(8,6));
plt.style.use('seaborn');
plt.scatter(lat, long, c=np.log10(population), cmap='viridis', s=area, linewidths=0, alpha=0.5);
plt.axis('equal');
plt.xlabel('longlatitude');
plt.ylabel('latitude');
plt.title('California Cities');
plt.colorbar(label='log$_{10}$(population)');
plt.clim(3,7);


# Create a legend for cities's sizes
area_range = [50,100,300,500]
for area in area_range:
  plt.scatter([], [],c='k',alpha=0.3, s=area, label=str(area)+'km$^2$');
plt.legend(scatterpoints = 1,labelspacing = 1, title='City Area');
```
